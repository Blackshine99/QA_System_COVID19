```python
import json
```

```python
with open("questions-pt1.json") as f:
    data = json.load(f)
```

```python
print(json.dumps(data, indent=4))
```

```python
c=0
for category in data['categories']:
    print(category['name'])
    c+=1
```

```python
questions = []
answers = []
for i in range(c):
    for q in data['categories'][i]['sub_categories']:
        print(q['nq_name'])
        questions.append(q['nq_name'])
        answer = ""
        for ans in q['answers']:
            answer+=ans['title'] + " : " + ans['exact_answer'] + "\n"
        answers.append(answer)    
```

```python
for ques, ans in zip(questions, answers):
    print("Question: ", ques)
    print("Answer:\n", ans)
    print("\n\n--------\n\n")
```

```python
import pandas as pd
data = {'question': questions,
       'answer': answers}
df = pd.DataFrame(data)
```
